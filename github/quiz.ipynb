4.1 Что нужно сделать чтобы открыть пул реквест? (Рисунок 90)

# Для открытия Pull Request нужно внести изменения в отдельную ветку, создать commit, отправить ветку в удалённый репозиторий и затем на GitHub нажать "New pull request", выбрать нужные ветки и создать запрос на слияние.

4.2 Что нужно сделать Если ваш Форк устарел?

# Чтобы обновить устаревший Fork, нужно получить изменения из оригинального репозитория (git fetch upstream), слить их с текущей веткой (git merge upstream/main) и отправить обновления в свой Fork (git push origin main).

4.3 Что нужно сделать если в пул реквесте имеются конфликты слияния (Рисунок 96)

# При возникновении конфликтов слияния нужно скачать изменения, обновить свою ветку относительно целевой ветки, вручную разрешить конфликты в файлах, сделать новый commit и отправить изменения в Fork. После этого Pull Request обновится.

5 Что нужно сделать Для добавления отрывка кода в комментарии к ишьюсу? (Рисунок 104)

# Чтобы добавить отрывок кода в комментарий к Issue, нужно обрамить код тремя обратными кавычками (```) до и после кода. После первых трёх кавычек можно указать язык программирования (например, java или python), тогда GitHub добавит подсветку синтаксиса.

# Картинку в Issue на GitHub можно вставить, перетащив изображение мышкой (drag & drop) прямо в поле комментария или нажав "selecting them" и выбрав файл с компьютера. После загрузки GitHub автоматически создаст Markdown-ссылку на изображение.
