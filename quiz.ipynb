23.3) Различие между стэшем и коммитом.  
Коммит — это сохранение законченных изменений в историю Git. Его можно потом отправить на GitHub через push.
Stash — это временный буфер для незакоммиченных изменений. Он нужен, например, когда коллега изменил тот же файл и нужно сначала сделать pull, а свои текущие изменения временно спрятать.
  
  Когда лучше сохранить изменения в стэше, а когда коммитить.
    В stash лучше сохранять изменения, когда работа ещё не закончена, коммит делать рано или нужно быстро переключиться на другую ветку / принять pull.

    Коммит лучше делать, когда часть работы закончена, изменения понятны и их можно сохранить в историю и отправить на GitHub.

  Удалить один stash: git stash drop stash@{0}
  Удалить все stash: git stash clear

  Краткие сценарии, где стэш помогает.
  Stash помогает, когда я ещё не закончил изменения, но нужно сделать pull, потому что в GitHub уже появились изменения коллеги в том же файле.

  Также stash нужен, когда надо переключиться на другую ветку, а текущие изменения пока не готовы для коммита. Сначала прячу их в stash, потом переключаюсь на другую ветку, а позже восстанавливаю свои изменения.

24) Где посмотреть что есть конфликт в файлах? 
Конфликт можно увидеть в VS Code или Cursor: файл появится в разделе Changes как конфликтный, а внутри файла будут кнопки `Accept Current Change`, `Accept Incoming Change` и `Accept Both Changes`.
24.1) Когда он появляется? 
Конфликт появляется, когда в одном и том же месте одного файла были сделаны разные изменения: например, ты изменил файл локально, а коллега изменил те же строки в GitHub. Обычно это может случиться после `pull`, `merge` или восстановления stash.
25) Как решить конфликт в файлах?
В VSC откроем проблемный файл. зелененький текст это с гитхаба. синий это локальные изменения. Accept Current Change принять текущие изменения. Accept Incoming Change принять из вне изменения. Accept Both Changes принять оба. Либо вручную в файле удалить что ненужно и потом нажатьResolve in Merge Editor. Потом complete merge.

27) Если мы работаем в одном файле, можно ли принять pull после того как вы спрячете в стэш свои изменения? да.Сначала нужно сохранить свои незакоммиченные изменения в stash, потом сделать pull и принять изменения с GitHub.
27.1) Что может произойти когда stash восстановите после принятия pull?
Может появиться конфликт в файле, если мои изменения из stash и изменения, которые пришли через pull, затрагивают одни и те же строки. Если изменения были в разных файлах или в разных местах файла, то конфликт обычно не появится.
28) Сколько способов решения конфликтов было показано в видео? Напишите ЧИСЛО и укажите их способы.
4 способа:

1. `Accept Current Change` — оставить текущие изменения.
2. `Accept Incoming Change` — принять изменения, пришедшие с GitHub.
3. `Accept Both Changes` — оставить оба варианта.
4. Вручную отредактировать файл: удалить ненужный текст и оставить правильный вариант.
29) Что делает кнопка complete merge?
Кнопка `Complete Merge` завершает слияние после решения конфликтов и сохраняет выбранный итоговый вариант файла.

Пакетный менеджер — это программа для установки, обновления и удаления библиотек Python.

Для Data Science чаще удобнее conda, потому что он устанавливает не только Python-библиотеки, но и зависимости для работы с ними, а также создаёт отдельные окружения. Но pip вместе с виртуальным окружением тоже нормально использовать.
3) Почему расширение py лучше чем ipynb?
Файл `.py` удобнее для Git и проверки кода линтерами. В нём хранится только обычный Python-код и текстовые комментарии, поэтому изменения в коммитах легче читать и проверять.

Файл `.ipynb` внутри хранится как большой JSON-файл: там есть ячейки, результаты запусков и другая служебная информация. Из-за этого при сравнении версий могут появляться лишние изменения и конфликты.

Но `.ipynb` удобнее для обучения, экспериментов, графиков и запуска кода по ячейкам.
