23) Что такое stash?

Stash — это временный буфер, где хранятся незакоммиченные изменения.

Он нужен, например, если вы работаете с коллегой в одном файле и перед `pull` надо временно спрятать свои изменения.

23.1) Как сохранить stash?

Git-команда:

```bash
git stash save "NAME_STASH"
```

В VS Code:

```text
Ctrl + Shift + P → Git: Stash
```

23.2) Как восстановить stash?

В GitHub Desktop слева нажать `Stashed Changes`, потом `Restore`.

Git-команда:

```bash
git stash apply stash@{0}
```

23.3) Различие между stash и коммитом. Когда лучше сохранить изменения в stash, а когда коммитить?

Коммит — это сохранение законченных изменений в историю Git. Его можно потом отправить на GitHub через `push`.

Stash — это временный буфер для незакоммиченных изменений. Он нужен, например, когда коллега изменил тот же файл и нужно сначала сделать `pull`, а свои текущие изменения временно спрятать.

В stash лучше сохранять изменения, когда работа ещё не закончена, коммит делать рано или нужно быстро переключиться на другую ветку / принять `pull`.

Коммит лучше делать, когда часть работы закончена, изменения понятны и их можно сохранить в историю и отправить на GitHub.

23.4) Как просмотреть список сохранённых stash?

```bash
git stash list
```

23.5) Как удалить stash?

Удалить один stash:

```bash
git stash drop stash@{0}
```

Удалить все stash:

```bash
git stash clear
```

23.6) Практические примеры использования stash.

Stash помогает, когда я ещё не закончил изменения, но нужно сделать `pull`, потому что на GitHub уже появились изменения коллеги в том же файле.

Также stash нужен, когда надо переключиться на другую ветку, а текущие изменения пока не готовы для коммита. Сначала прячу их в stash, потом переключаюсь на другую ветку, а позже восстанавливаю свои изменения.

24) Где посмотреть, что есть конфликт в файлах?

Конфликт можно увидеть в VS Code или Cursor: файл появится в разделе Changes как конфликтный, а внутри файла будут кнопки:

- `Accept Current Change`
- `Accept Incoming Change`
- `Accept Both Changes`

24.1) Когда он появляется?

Конфликт появляется, когда в одном и том же месте одного файла были сделаны разные изменения: например, ты изменил файл локально, а коллега изменил те же строки в GitHub.

Обычно это может случиться после `pull`, `merge` или восстановления stash.

25) Как решить конфликт в файлах?

В VS Code откроем проблемный файл.

Зелёный текст — то, что пришло с GitHub. Синий — локальные изменения.

- `Accept Current Change` — принять текущие изменения.
- `Accept Incoming Change` — принять внешние изменения.
- `Accept Both Changes` — принять оба варианта.
- Либо вручную удалить ненужный текст и оставить правильный вариант.

Потом нажать `Resolve in Merge Editor`, затем `Complete Merge`.

26) Напишите правильное утверждение.

Зелёное — то, что пришло с GitHub, а синее — локальные изменения.

27) Если мы работаем в одном файле, можно ли принять pull после того, как вы спрячете в stash свои изменения?

Да.

Сначала нужно сохранить свои незакоммиченные изменения в stash, потом сделать `pull` и принять изменения с GitHub.

27.1) Что может произойти, когда stash восстановите после принятия pull?

Может появиться конфликт в файле, если мои изменения из stash и изменения, которые пришли через `pull`, затрагивают одни и те же строки.

Если изменения были в разных файлах или в разных местах файла, то конфликт обычно не появится.

28) Сколько способов решения конфликтов было показано в видео? Напишите ЧИСЛО и укажите их способы.

4 способа:

1. `Accept Current Change` — оставить текущие изменения.
2. `Accept Incoming Change` — принять изменения, пришедшие с GitHub.
3. `Accept Both Changes` — оставить оба варианта.
4. Вручную отредактировать файл: удалить ненужный текст и оставить правильный вариант.

29) Что делает кнопка Complete Merge?

Кнопка `Complete Merge` завершает слияние после решения конфликтов и сохраняет выбранный итоговый вариант файла.

# Дополнительные вопросы

1) Какая команда конвертирует файл в py из ipynb?

```bash
jupytext --to py название_файла.ipynb
```

2) Что такое пакетный менеджер? Вы пользуетесь пакетным менеджером conda или pip? Какой лучше использовать для Data Science?

Пакетный менеджер — это программа для установки, обновления и удаления библиотек Python.

Я пользуюсь `pip`, например:

```bash
pip install pandas
```

Для Data Science часто удобнее `conda`, потому что он устанавливает не только Python-библиотеки, но и зависимости для работы с ними, а также создаёт отдельные окружения.

Но `pip` вместе с виртуальным окружением тоже нормально использовать.

3) Почему расширение py лучше, чем ipynb?

Файл `.py` удобнее для Git и проверки кода линтерами. В нём хранится только обычный Python-код и текстовые комментарии, поэтому изменения в коммитах легче читать и проверять.

Файл `.ipynb` внутри хранится как большой JSON-файл: там есть ячейки, результаты запусков и другая служебная информация. Из-за этого при сравнении версий могут появляться лишние изменения и конфликты.

Но `.ipynb` удобнее для обучения, экспериментов, графиков и запуска кода по ячейкам.

4) Что такое PEP 8?

PEP 8 — это рекомендации по стилю написания Python-кода.

В них указано, как называть переменные, делать отступы, переносить длинные строки и оформлять код, чтобы он был понятным и одинаковым у разных программистов.

```text
https://peps.python.org/pep-0008/
```

4.1) Линтеры проверяют PEP 8?

Да.

Например, `Flake8` проверяет многие правила PEP 8: отступы, длину строк, пробелы и другие ошибки стиля.

`Pylint` тоже может проверять стиль кода, но у него есть и другие проверки.

4.2) Какая нотация используется для создания переменных?

Для переменных в Python используется `snake_case`.

4.3) Может ли переменная состоять из одной буквы, например underscore `_`?

Да.

4.4) Зачем и где мы используем underscore `_`?

Underscore `_` используют в названиях переменных и функций, чтобы разделять слова в стиле `snake_case`.

Также `_` используют как переменную, значение которой не нужно, например в цикле:

```python
for _ in range(5):
    print("Привет")
```

4.5) По PEP 8 допустима переменная в одну букву?

Да, допустима, например `i`, `j` или `x` для коротких локальных переменных и счётчиков в цикле.

Но PEP 8 не рекомендует использовать `l`, `O` и `I`, потому что их легко перепутать с цифрами.
